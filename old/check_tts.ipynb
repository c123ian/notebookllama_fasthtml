```python
speaker1_description = """
Laura's voice is expressive and dramatic in delivery, speaking at a moderately fast pace with a very close recording that almost has no background noise.
"""
speaker2_description = """ 
Gary's voice is expressive and dramatic in delivery, speaking at a slow pace with a very close recording that almost has no background noise.
"""


preprocess_text = """
Exactly! And the distillation part is where you take a LARGE-model,and compress-it down into a smaller, more efficient model that can run on devices with limited resources.
"""

from scipy.io import wavfile
from pydub import AudioSegment
import io

```

```python
# New function to split TTS text into chunks of a target character length (adjust as needed)
def create_tts_chunks(text, target_length=400):
    words = text.split()
    chunks = []
    current_chunk = []
    current_length = 0
    for word in words:
        word_len = len(word) + 1
        if current_length + word_len > target_length and current_chunk:
            chunks.append(' '.join(current_chunk))
            current_chunk = [word]
            current_length = word_len
        else:
            current_chunk.append(word)
            current_length += word_len
    if current_chunk:
        chunks.append(' '.join(current_chunk))
    return chunks



def numpy_to_audio_segment(audio_arr, sampling_rate):
    audio_int16 = (audio_arr * 32767).astype(np.int16)
    byte_io = io.BytesIO()
    wavfile.write(byte_io, sampling_rate, audio_int16)
    byte_io.seek(0)
    return AudioSegment.from_wav(byte_io)
```

```python
def generate_speaker_audio(text, speaker="Speaker 1"):
        if not isinstance(text, str):
            text = str(text)
        text = preprocess_text(text)
        desc = speaker1_description if speaker == "Speaker 1" else speaker2_description
        # Chunk the text into parts that are short enough (e.g. 400 characters)
        tts_chunks = create_tts_chunks(text, target_length=400)
        audio_segments = []
        for chunk in tts_chunks:
            print(f"🎤 Generating TTS audio for {speaker} chunk: {len(chunk)} chars")
            input_data = tokenizer(desc, return_tensors="pt", truncation=True, max_length=512, padding="max_length").to(device)
            prompt_data = tokenizer(chunk, return_tensors="pt", truncation=True, max_length=512, padding="max_length").to(device)
            try:
                generation = model.generate(
                    input_ids=input_data["input_ids"],
                    attention_mask=input_data["attention_mask"],
                    prompt_input_ids=prompt_data["input_ids"],
                    temperature=0.9
                )
                generation = generation.cpu().float()
                audio_arr = generation.numpy().squeeze()
                if np.isnan(audio_arr).any() or np.isinf(audio_arr).any():
                    raise ValueError("Invalid audio generation output")
                audio_segments.append(audio_arr)
            except Exception as e:
                print(f"Error generating audio for {speaker} chunk: {str(e)}")
                # Use 1 second of silence if generation fails
                audio_segments.append(np.zeros(int(model.config.sampling_rate)))
        # Concatenate the generated audio chunks together
        if audio_segments:
            # For simplicity, we use the first chunk's sampling rate
            sampling_rate = model.config.sampling_rate
            # Convert each numpy audio array to an AudioSegment and then append
            segments = [numpy_to_audio_segment(seg, sampling_rate) for seg in audio_segments]
            combined = segments[0]
            for seg in segments[1:]:
                combined = combined.append(seg, crossfade=100)
            # Convert combined audio back to numpy array (for consistency with rest of pipeline)
            # Here we export to a temporary WAV in memory and read it back
            buf = io.BytesIO()
            combined.export(buf, format="wav")
            buf.seek(0)
            rate, final_audio_arr = wavfile.read(buf)
            return final_audio_arr, rate
        else:
            return np.zeros(int(model.config.sampling_rate)), model.config.sampling_rate

def concatenate_audio_segments(segments, sampling_rates):
        final_audio = None
        for segment, rate in zip(segments, sampling_rates):
            audio_segment = numpy_to_audio_segment(segment, rate)
            if final_audio is None:
                final_audio = audio_segment
            else:
                final_audio = final_audio.append(audio_segment, crossfade=100)
        return final_audio
```
